### Distribuição de variáveis relevantes:

```python
print(cadastro_tratamento_de_agua.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Gráfico de Cadastro de Tratamento de Água
# Usando a coluna "Etapa desinfecção" como exemplo
cadastro_tratamento_de_agua['tipo_tratamento'] = cadastro_tratamento_de_agua['Etapa desinfecção'].fillna('Não especificado')
sns.countplot(x="tipo_tratamento", data=cadastro_tratamento_de_agua)
plt.title("Distribuição de Tipos de Tratamento de Água")
plt.xticks(rotation=45)  # Rotacionar os rótulos para melhor leitura
plt.show()
```

```python
# 2. Gráfico de Estado Nutricional (SISVAN)
# Usando a coluna "faixa_etaria" como exemplo (ajuste se o nome da coluna for diferente)
sns.countplot(x='DS_FASE_VIDA', data=sisvan_estado_nutricional_2023)
plt.title("Distribuição por Fase de Vida no SISVAN")
plt.xlabel("Fase de Vida")
plt.ylabel("Frequência")
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Gráfico de Municípios por Região
# Usando a coluna "Região Geográfica" como exemplo
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

estados = pd.read_csv("estados.csv")

municipios_com_regiao = municipios.merge(estados[['codigo_uf', 'regiao']], on='codigo_uf', how='left')

# Verificar se a coluna 'regiao' foi adicionada corretamente
print(municipios_com_regiao.head())

# Gerar o gráfico de distribuição de municípios por região geográfica
sns.countplot(x="regiao", data=municipios_com_regiao, order=municipios_com_regiao["regiao"].value_counts().index)
plt.title("Distribuição de Municípios por Região Geográfica")
plt.xticks(rotation=45)  # Rotacionar os rótulos para melhor leitura
plt.xlabel("Região Geográfica")
plt.ylabel("Número de Municípios")
plt.show()

```
